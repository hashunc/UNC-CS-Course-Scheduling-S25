```python
import sqlite3
import requests as rq
```

```python
def fetch_qualified_courses_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT * FROM QualifiedCourses"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

qcourses = fetch_qualified_courses_as_dict()
qcourses
```

```python
def fetch_profs_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT DISTINCT Prof FROM Availability"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

profs = fetch_profs_as_dict()
new_profs = []
for prof in profs:
    new_profs.append(prof['Prof'])
# profs = list(map(lambda x: x['Prof'], profs))
new_profs
```

```python
professors_json = {}
for prof in new_profs:
    professors_json[prof] = {
        'qualified_courses': [],
        'availability': [],
        'max_classes': 0
    }

professors_json
```

```python
for qcourse in qcourses:
    professor = qcourse['Prof']
    professors_json[professor]['qualified_courses'].append(qcourse['Course'])

professors_json
```

```python
def fetch_availabilities_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT * FROM Availability"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

availabilities = fetch_availabilities_as_dict()
availabilities
```

```python
def fetch__rooms_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT Room FROM Rooms"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

rooms = fetch__rooms_as_dict()
rlist = []
for room in rooms:
    rlist.append(room['Room'])
rlist

#rooms = list(map(lambda x: x['Room'], rooms))
```

```python
rooms_json = {}
for room in rlist:
    rooms_json[room] = {
        'capacity': 0,
    }
rooms_json
```

```python
def fetch__capacities_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT SeatCapacity FROM Rooms"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

capacities = fetch__capacities_as_dict()
c_list = []
for capacity in capacities:
    c_list.append(capacity['SeatCapacity'])

c_list
```

```python

for i, x in enumerate(rlist):  
    if x in rooms_json: 
        rooms_json[x]['capacity'] = c_list[i]

rooms_json
```

```python
def fetch__courses_as_dict():
    connection = sqlite3.connect("database.db")
    connection.row_factory = sqlite3.Row

    cursor = connection.cursor()
    try:
        query = "SELECT * FROM CoursesAndSections"
        cursor.execute(query)
        rows = cursor.fetchall()
        result = [dict(row) for row in rows]
        return result
    except sqlite3.Error as error:
        print(f"Error occurred: {error}")
        return []
    finally:
        connection.close()
        print("Database connection closed.")

courses = fetch__courses_as_dict()
new_courses = []
for course in courses:
    new_courses.append(course['Course'])
new_courses
```

```python
courses_json = {}
for course in new_courses:
    courses_json[course] = {
        'title': '',
        'sections': []
    }

courses_json
```

```python
courses_json = {}

for course in courses:
    course_name = course['Course']
    title = course['Title']
    section_number = int(course['Section']) if course['Section'].isdigit() else course['Section']
    seat_capacity = course['SeatCapacity']
    if course_name not in courses_json:
        courses_json[course_name] = {
            "title": title,
            "sections": []
        }
    courses_json[course_name]["sections"].append({
        "section_number": section_number,
        "seat_capacity": seat_capacity
    })

courses_json
```
